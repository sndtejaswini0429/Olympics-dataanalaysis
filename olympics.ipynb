```python
import pandas as pd
df = pd.read_csv('summer.csv')
print(df.head())
print(df.tail())
```

## **1. In how many cities Summer Olympics is held so far?**

```python
var = df["City"].value_counts()
print(var)
print("number of cities in Olympics held so far is:" ,len(var))
```


## **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
v = df['Sport'].loc[df['Medal'] == 'Gold']
#print(v)
#print(v.max())
v.value_counts()
#v.value_counts().head().plot.bar()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
sports = ['Aquatics' , 'Athletics' , 'Rowing' , 'Gymnastics' , 'Fencing' ]
g_medal = [1421,1215,890,820,552]
plt.bar(sports,g_medal)
plt.xlabel('sports')
plt.ylabel('no of gold medals')
plt.title('top5 sports with more gold medals')
plt.tight_layout()
plt.show()
```

## **6. In which year India won first Gold Medal in Summer Olympics?**
